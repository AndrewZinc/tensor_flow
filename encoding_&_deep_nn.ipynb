```python
# Import the dependencies
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
import tensorflow as tf

# Read the input data
hr_df = pd.read_csv('./Data/HR-Employee-Attrition.csv')
hr_df
```

```python
# Generate a categorical variable list
attrit_cat = hr_df.dtypes[hr_df.dtypes=='object'].index.tolist()
```

```python
# Check the number of unique values in each column
hr_df[attrit_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform using the categorical variable list
enc_df = pd.DataFrame(enc.fit_transform(hr_df[attrit_cat]))

# Add the encoded variables to the DataFrame
enc_df.columns = enc.get_feature_names(attrit_cat)
enc_df
```

```python
# Merge the encoded features and drop the originals
hr_df = hr_df.merge(enc_df,left_index=True, right_index=True)

hr_df = hr_df.drop(attrit_cat,1)
hr_df
```

```python
# split the processed data into our features and target arrays
y = hr_df['Attrition_Yes'].values

x = hr_df.drop(['Attrition_Yes','Attrition_No'],1).values

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=78)
```

```python
# Instantiate StandardScaler
scaler = StandardScaler()

# Fit the scaler
x_scaler = scaler.fit(x_train)

# Scale the data
x_train_scaled = x_scaler.transform(x_train)
x_test_scaled = x_scaler.transform(x_test)
```

```python
# Define the model - deep neural net
n_input_features = len(x_train[0])
hidden_nodes_l1 = 8
hidden_nodes_l2 = 5

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_l1, input_dim=n_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_l2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the model structure
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(x_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model with the test data

model_loss, model_accuracy = nn.evaluate(x_test_scaled, y_test, verbose=2)

print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
